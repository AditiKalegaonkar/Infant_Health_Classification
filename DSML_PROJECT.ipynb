Encoding the Data

```python
# Converting the string values to comparable float and int values to train the model and do the predictions
from sklearn.preprocessing import OrdinalEncoder
# Select categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
# Initialize ordinal encoder
encoder = OrdinalEncoder()
# Fit and transform categorical columns
df[cat_cols] = encoder.fit_transform(df[cat_cols])

df
```

Cross Validation

```python
x = df.drop(["Sick"], axis = 1)
y = df["Sick"]                 
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 0)
```

```python
print("Original class distribution:\n", y_train.value_counts())
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
print("After SMOTE:\n", y_train.value_counts())
```

Building the Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(
    n_estimators=500,         
    criterion='entropy',      
    max_depth=5,              
    min_samples_split=3,      
    min_samples_leaf=2,       
    max_features='sqrt',      
    bootstrap=True,          
    class_weight='balanced',  
    random_state=42
)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test) 
```

```python
df1 = pd.DataFrame({"Actual": y_test,"Predicted": y_pred})
df1
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
RFC = accuracy_score(y_test,y_pred)
```

Building the model for KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 20)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
KNN = accuracy_score(y_test,y_pred)
```

Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sfs = SequentialFeatureSelector(knn,direction = "forward") 
```

```python
sfs.fit(x_train,y_train)
```

```python
sfs.get_support()
```

```python
x = x.iloc[:,sfs.get_support()]
```

```python
knn.fit(x,y)
```

```python
y_pred = knn.predict(x)
accuracy_score(y,y_pred)
```

```python
KNN1 = knn.score(x,y)
```

Using hyperparameter tuning for increasing accuracy

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors': [20, 30, 40, 50],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
```

```python
grid =  GridSearchCV(estimator =KNeighborsClassifier(),
                    param_grid = params,
                    scoring = "accuracy",
                    cv = 5)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
```

```python
y_pred = grid.predict(x_test)
```

```python
KNN2 = accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred));
```
